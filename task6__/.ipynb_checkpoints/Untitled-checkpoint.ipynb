```python
import gym
import numpy as np
import random
```

```python
env = gym.make('Taxi-v3')
```

```python
q_table = np.zeros([env.observation_space.n,env.action_space.n])
```

```python
alpha = 0.1
gamma = 0.9
epsilon = 0.1
```

```python
rewards_list = []
dropouts_list = []
```

```python
episode_number = 10000
```

```python
for i in range(1, episode_number):
    state = env.reset()
    reward_count = 0
    dropouts = 0
    while True:
        if random.uniform(0,1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state])
     
        next_state, reward, done, _ = env.step(action)
        old_value = q_table[state, action]
        next_max = np.max(q_table[next_state])
        next_value = (1-alpha) * old_value * (reward + gamma * next_max)
        q_table[state, action] = next_value
        state = next_state
        if reward == -10:
            dropouts += 1
        if done:
            break
        reward_count += reward
        
        if i % 10 == 0:
            dropouts_list.append(dropouts)
            rewards_list.append(reward_count)
            print("Episoda : {}, Reward : {}, Wrong Dropout : {}".format(i, reward_count, dropouts))
```
